#### 本篇笔记使用python实现基于RBF核的主成分分析(PCA)，借助于库SciPy库和NumPy库。

使用核主成分分析可以将非线性可分的数据转换到一个线性可分的低维子空间上。方法是通过非线性的核函数，将原始的数据映射到维度更高的k维子空间。可以将核函数理解为通过两个向量点积来度量向量间的相似度的函数。

可以通过以下三个步骤来实现一个基于RBF核的PCA。

##### <font color=red>1)计算径向基核函数:</font>
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\exp \left(-\gamma\left\|\boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}\right\|^{2}\right)
$$

需要计算任意两个样本对之间的值：

$$
\boldsymbol{K}=\left[\begin{array}{cccc}{\boldsymbol{K}\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(1)}\right)} & {\boldsymbol{K}\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(2)}\right)} & {\cdots} & {\boldsymbol{K}\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(n)}\right)} \\ {\boldsymbol{K}\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(1)}\right)} & {\boldsymbol{K}\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(2)}\right)} & {\cdots} & {\boldsymbol{K}\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(n)}\right)} \\ {\vdots} & {\vdots} & {\vdots} & {\vdots} \\ {\kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(1)}\right)} & {\boldsymbol{K}\left(\boldsymbol{x}^{(d)}, \boldsymbol{x}^{(2)}\right)} & {\cdots} & {\boldsymbol{K}\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(n)}\right)}\end{array}\right]
$$

##### <font color=red>2)通过以下公式进行计算，使得核矩阵k更为聚集：</font>
$$
\boldsymbol{K}^{\prime}=\boldsymbol{K}-1_{n} \boldsymbol{K}-\boldsymbol{K} 1_{n}+1_{n} \boldsymbol{K} 1_{n}
$$

其中$l_{n}$是一个nxn维的矩阵(与核矩阵维度相同)，其所有值均为$\frac{1}{n}$。

##### <font color=red>3)将聚集后的核矩阵的特征值降序排列，选择前k个特征值所对应的特征向量。</font>
与标准PCA不同的是，这里的特征向量不是主成分轴，而是将样本映射到这些轴上。

###### 1.生成半月形数据

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=100, random_state=123)
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o',alpha=0.5)
plt.show()
```

###### 2.使用标准PCA对数据进行降维

```python
from sklearn.decomposition import PCA
import numpy as np

scikit_pca = PCA(n_components=2)

X_spca = scikit_pca.fit_transform(X)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))

ax[0].scatter(X_spca[y==0, 0], X_spca[y==0, 1], color='red', marker='o', alpha=0.5)
ax[0].scatter(X_spca[y==1, 0], X_spca[y==1, 1], color='blue', marker='o', alpha=0.5)
ax[1].scatter(X_spca[y==0, 0], np.zeros((50, 1)) + 0.02, color='red', marker='^', alpha=0.5)
ax[1].scatter(X_spca[y==1, 0], np.zeros((50, 1)) + 0.02, color='blue', marker='o', alpha=0.5)

ax[0].set_xlabel('PC 1')
ax[0].set_ylabel('PC 2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC 1')

plt.show()
```

注：标准PCA并没有把原始数据变成线性可分的

###### 3.实现核PCA函数，并且将其运用于半月形数据的降维

```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np


def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.
    Parameters
    -------------
    X: {Numpy ndarray}, shape=[n_sanples, n_features]

    gamma: float
        Tuning parameter of the RBF kernel

    n_components: int
        Number of the principal components to return

    Returns
    -------------
    X_pc: {Numpy ndarray}, shape=[n_samples, k_features]
        Projected dataset
    """

    # Calculate pairwise squared Euclidean distances
    # in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')

    # Convert pairwise distances into a square matrix
    mat_sq_dists = squareform(sq_dists)

    # Compute the symmetric kernel matrix.
    K = exp(-gamma * mat_sq_dists)

    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Obtaining eigenpairs from the centered kernel matrix
    # numpy.eigh returns them in sorted order

    eigcals, eigvecs = eigh(K)
    # Collet the top k eigenvectors(projected samples)
    X_pc = np.column_stack([eigvecs[:, -i] for i in range(1, n_components + 1)])
    return X_pc
```

```python
from matplotlib.ticker import FormatStrFormatter
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np

X, y = make_moons(n_samples=100, random_state=123)
X_kpca = rbf_kernel_pca(X, gamma=15, n_components=2)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
ax[0].scatter(X_kpca[y == 0, 0], X_kpca[y == 0, 1], color='red', marker='^', alpha=0.5)
ax[0].scatter(X_kpca[y == 1, 0], X_kpca[y == 1, 1], color='blue', marker='o', alpha=0.5)
ax[1].scatter(X_kpca[y == 0, 0], np.zeros((50, 1)) - 0.02, color='red', marker='^', alpha=0.5)
ax[1].scatter(X_kpca[y == 1, 0], np.zeros((50, 1)) - 0.02, color='blue', marker='o', alpha=0.5)

ax[0].set_xlabel('PC 1')
ax[0].set_ylabel('PC 2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC 1')

ax[0].xaxis.set_major_formatter(FormatStrFormatter('%0.1f'))
ax[1].xaxis.set_major_formatter(FormatStrFormatter('%0.1f'))

plt.show()
```

注：经过核PCA处理后，原始的非线性的半月形数据变成线性可分的

###### 5.映射新的数据点

在核PCA中，如果想把新的数据集映射到主成分轴上，可以使用核函数的技巧来跳过构造映射矩阵的步骤。按照此思路可得到如下公式：

$$
\begin{aligned} \phi\left(\boldsymbol{x}^{\prime}\right)^{\mathrm{T}} \boldsymbol{v} &=\sum_{i} a^{(i)} \boldsymbol{\phi}\left(\boldsymbol{x}^{\prime}\right)^{\mathrm{T}} \boldsymbol{\phi}\left(\boldsymbol{x}^{(i)}\right) \\ &=\sum_{i} a^{(i)} k\left(\boldsymbol{x}^{\prime}, \boldsymbol{x}^{(i)}\right)^{\mathrm{T}} \end{aligned}
$$

其中a是核矩阵K的特征向量，完成上述计算后，还需要通过特征向量对应的特征值来对其进行归一化的处理。

修改上面实现过的rbf_kernel_pcah函数，返回核矩阵的特征值和特征向量。

```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np


def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.
    Parameters
    -------------
    X: {Numpy ndarray}, shape=[n_sanples, n_features]

    gamma: float
        Tuning parameter of the RBF kernel

    n_components: int
        Number of the principal components to return

    Returns
    -------------
    X_pc: {Numpy ndarray}, shape=[n_samples, k_features]
        Projected dataset
    """

    # Calculate pairwise squared Euclidean distances
    # in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean') # 在MxN维的数据集中计算两两的欧式矩阵

    # Convert pairwise distances into a square matrix
    mat_sq_dists = squareform(sq_dists) # 转换为平方距离矩阵

    # Compute the symmetric kernel matrix.
    K = exp(-gamma * mat_sq_dists) # 计算核矩阵

    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Obtaining eigenpairs from the centered kernel matrix
    # numpy.eigh returns them in sorted order
    eigvals, eigvecs = eigh(K) # eigvecs.shape=(100, 100)

    # Collet the top k eigenvectors(projected samples)
    alphas = np.column_stack([eigvecs[:, -i] for i in range(1, n_components + 1)]) # 返回的特征值是排好序的，取k个最大的特征值对应的特征向量
    
    # Collet the corresponding eigenvalues
    lambdas = [eigvals[-i] for i in range(1, n_components+1)]
    return alphas, lambdas
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=100, random_state=123) # X.shape=(100, 2)
alphas, lambdas = rbf_kernel_pca(X, gamma=15, n_components=1) # 修改以后的函数返回核矩阵的特征值和特征向量
                                                              # alphas.shape=(100, 1)
```

```python
x_new = X[25]
x_new
```

```python
x_proj = alphas[25] # original projection
x_proj
```

```python
def project_x(x_new, X, gamma, alphas, lambdas):
    # 将数据映射到主成分轴上
    pair_dist = np.array([np.sum((x_new - row) ** 2) for row in X])    
    k = np.exp(-gamma * pair_dist)
    new = k.dot(alphas) / lambdas
    return new
```

```python
x_reproj = project_x(x_new, X, gamma=15, alphas=alphas, lambdas=lambdas)
x_reproj
```

对比原始的映射数据以及project_x函数处理以后得到的数据可知，两个值是相同的，所以新的数据集(非训练集中的数据)可以直接通过调用project_x方法直接映射到主成分轴上。

```python
import matplotlib.pyplot as plt

plt.scatter(alphas[y==0, 0], np.zeros((50)), color='red', marker='^', alpha=0.5)
plt.scatter(alphas[y==1, 0], np.zeros((50)), color='blue', marker='o', alpha=0.5)
plt.scatter(x_proj, 0, color='black', label='original projection of ponit X[25]', marker='^', s=100)
plt.scatter(x_reproj, 0, color='green', label='remapped point X[25]', marker='x', s=500)
plt.legend(scatterpoints=1)
plt.show()
```

##### 6.scikit-learn中的的核主成分分析

```python
from sklearn.decomposition import KernelPCA

X, y = make_moons(n_samples=100, random_state=123)
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)

X_skernpca = scikit_kpca.fit_transform(X)

plt.scatter(X_skernpca[y==0, 0], X_skernpca[y==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], X_skernpca[y==1, 1], color='blue', marker='^', alpha=0.5)

plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

总结：至此已经学习了三种基于特征提取的基本的数据降维技术：标准PCA、LDA以及核PCA。PCA是一种无监督的技术，可以在忽略类标的情况下，将数据映射到低维的特征子空间，并使得新的特征空间上数据沿着正交的特征坐标方向有最大方差。LDA是一种有监督的降维技术，已可以发现最优分类的特征子空间为目标。也就是要将数据在低维度上进行投影，尽可能使得同一个类别的投影点尽可能地接近，而不同类别的投影点离得尽可能远。而核PCA则将非线性的数据集映射到一个低维的特征空间上，使得数据线性可分。
