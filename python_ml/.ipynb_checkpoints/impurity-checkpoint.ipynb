```python
import matplotlib.pyplot as plt
import numpy as np

def gini(p):
    return (p)*(1-(p)) + (1-p)*(1-(1-p))
def entropy(p):
    return -p*np.log2(p) - (1-p)*np.log2((1-p))
def error(p):
    return 1-np.max([p, 1-p])


x = np.arange(0.0, 1.0, 0.01) # 生成一个np数组
ent = [entropy(p) if p != 0 else None for p in x] # 计算熵，得到的仍是一个np数组
sc_ent = [e*0.5 if e else None for e in ent]  # 将熵进行压缩
err = [error(i) for i in x]
fig = plt.figure() # 生成画布
ax = plt.subplot(111) # 取子图
for i, lab, ls, c, in zip([ent, sc_ent, gini(x), err], 
                          ['Entropy', 'Entropy(scaled)', 'Gini Impurity', 'Misclassification Error'],
                          ['-', '-', '--', '-.'],
                          ['black', 'lightgray', 'red', 'green', 'cyan']):
    line = ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c) 
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--') # 画出y=0.5的横线
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--') # 画出y=1.0的横线
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```
